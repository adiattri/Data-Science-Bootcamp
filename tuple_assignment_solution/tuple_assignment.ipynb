```python
def min_in_tuple(tpl):
    print(min(tpl))

def max_in_tuple(tpl):
    print(max(tpl))

def sum_in_tuple(tpl):
    print(sum(tpl))

sample=(1,2,3)

min_in_tuple(sample)

max_in_tuple(sample)

sum_in_tuple(sample)
```
